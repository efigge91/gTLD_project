```python
import json
import pprint
import pandas as pd
from pandas.io.json import json_normalize
```

```python
with open('json_files/tlds.json') as json_data:
    tlds_json = json.load(json_data)
    
pprint.pprint(tlds_json)
```

```python
print (tlds_json['resData']['tlds'][0]['zone'])
```

```python
tld_list = []
count_list = []
url_list = []

for tld in tlds_json['resData']['tlds']:
    tld_list.append(tld['zone'])
    count_list.append(tld['total_domains'])
    url_list.append(tld['registry']['url'])


```

```python
tld_df = pd.DataFrame (tld_list, columns=['tlds'])
tld_df['count'] = count_list
tld_df['url']=url_list
tld_df
```

```python
with open('json_files/registries.json') as json_data:
    registries_json = json.load(json_data)
    
pprint.pprint(registries_json)
```

```python
registries_list = []
count_list = []
url_list = []

for registry in registries_json['resData']['registries']:
    registries_list.append(registry['name'])
    count_list.append(registry['total_domains'])
    url_list.append(registry['url'])
```

```python
registries_df = pd.DataFrame (registries_list, columns=['registries'])
registries_df['count'] = count_list
registries_df['url']=url_list
registries_df
```

```python
with open('json_files/registrars.json') as json_data:
    registrars_json = json.load(json_data)
    
pprint.pprint(registrars_json)
```

```python
registrars_list = []
count_list = []
url_list = []

for registrar in registrars_json['resData']['registrars']:
    registrars_list.append(registrar['name'])
    count_list.append(registrar['total_domains'])
    url_list.append(registrar['registrar_group']['url'])
```

```python
registrars_df = pd.DataFrame (registrars_list, columns=['registrars'])
registrars_df['count'] = count_list
registrars_df['url']=url_list
registrars_df
```

```python
tld_df.to_csv('csv_files/tld.csv')
registries_df.to_csv('csv_files/registries.csv')
registrars_df.to_csv('csv_files/registrars.csv')
```
